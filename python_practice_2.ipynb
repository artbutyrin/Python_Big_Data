8. Маючи рядок "My dog is crazy." перетворити його на список ["my", "dog", "is", "crazy"]

```python
# Початковий рядок
my_text = "My dog is crazy."  # Змінив назву змінної

# Перетворення
result = my_text.replace('.', '').lower().split()

# Вивід
print("Початковий рядок:", my_text)
print("Отриманий список:", result)
print("Кількість слів:", len(result))
```
